```python
import numpy as np
import pandas as pd

import re
import json
import os
import pathlib
```

```python
# COINBASE API

from coinbase.wallet.client import Client

with open("coinbase_api.token") as f:
    apifile = f.read().splitlines()
    api_key =  apifile[0].split(':')[1]
    api_secret = apifile[1].split(':')[1]
    print(f'api_key:{api_key}\napi_secret:{api_secret}')
    

client = Client(api_key=api_key, api_secret=api_secret, api_version='2021-10-16')


```

```python
# USER
user = client.get_current_user()
user_as_dict = json.loads(json.dumps(user))
user_id = user_as_dict['id']

```

```python
# ACCOUNTS
accounts = client.get_accounts() # 1 account for each type of CRYPTOCURRENCY (I guess)
#print(accounts)
primary = {}
primary['account'] = client.get_primary_account()
primary['id'] = primary['account']['id']


primary_account = client.get_primary_account()
#primary_account.set_primary()
```

```python
# ADDRESSES
addresses = client.get_addresses(primary_account['id'])
```
